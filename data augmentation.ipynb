```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen=ImageDataGenerator(rotation_range=40,
width_shift_range=0.2,
height_shift_range=0.2,
shear_range=0.2,
zoom_range=0.2,
horizontal_flip=True,
fill_mode='nearest')
```

```python
import os
from keras.preprocessing import image
import matplotlib.pyplot as plt

#image library me image processing utilities hote he

train_cats_dir='./cats_and_dogs_small/train/cats'
fnames = [os.path.join(train_cats_dir, fname) for fname in os.listdir(train_cats_dir)]

#selecting one image path to augment
image_path=fnames[0]

#image.load(path) to load image
img = image.load_img(image_path, target_size=(150, 150))


# Converts it to a Numpy array with shape (150, 150, 3)
x = image.img_to_array(img)

# Reshapes it to (1, 150, 150, 3)
x = x.reshape((1,) + x.shape)
```

```python
# datagen.flow() hame augmented data provide karta

i=0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1

    plt.show()
```
